```python
from rdkit import Chem
from rdkit.Chem import AllChem
from pathlib import Path

# 1) Описываем воду в виде SMILES (это H2O)
smiles = "O"

# 2) Создаём молекулу и добавляем явные H, чтобы получилось H2O
mol = Chem.MolFromSmiles(smiles)
mol = Chem.AddHs(mol)

# 3) Генерируем 3D-конформер
params = AllChem.ETKDGv3()
params.randomSeed = 0
ok = AllChem.EmbedMolecule(mol, params)

# 4) Опционально: немного оптимизируем геометрию
AllChem.UFFOptimizeMolecule(mol)

# 5) Сохраняем в SDF
out_path = Path("H2O_3D.sdf")
writer = Chem.SDWriter(str(out_path))
writer.write(mol)
writer.close()

print("Saved:", out_path.resolve())
print("Atoms:", mol.GetNumAtoms())
```
